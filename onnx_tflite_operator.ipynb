```python
import onnx
from onnx import shape_inference


model_path = "./0418/last_ckpt_opset11_0418.onnx"
model = onnx.load(model_path)

# Apply shape inference to the model
inferred_model = shape_inference.infer_shapes(model)

# Print the inferred shapes for each input and output
for inp in inferred_model.graph.input:
    print(inp.name, inp.type.tensor_type.shape.dim)
for out in inferred_model.graph.output:
    print(out.name, out.type.tensor_type.shape.dim)
```
